# Bibliotecas

```python
import plotly.figure_factory as ff
import pandas as pd
```

# Dado de entrada

```python
df = pd.read_excel("PIB_estados.xlsx")
df.head()
```

# Gráfico de compartimento hexagonal

```python
fig = ff.create_hexbin_mapbox(data_frame = df,
                             lat = "Latitude",
                             lon = "Longitude",
                             mapbox_style = "carto-positron",
                             min_count = 1,
                             show_original_data = True,
                             original_data_marker = dict(size = 4,
                                                        opacity = 0.6,
                                                        color = "black"),
                             zoom = 2,
                             color = "PIB per capita",
                             labels = {"color": "PIB per capita"},
                             nx_hexagon = 5,
                             opacity = 0.9,
                             color_continuous_scale = "Viridis")

fig.show()
```
